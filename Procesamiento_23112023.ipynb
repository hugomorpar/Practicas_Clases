```python
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(font_scale=1)
matplotlib.style.use('ggplot')
null_columns = titanic.columns[titanic.isnull().any()]
# pd.options.display.mpl_style = 'default'
labels = []
values = []
for col in null_columns:
    labels.append(col)
    values.append(titanic[col].isnull().sum())
ind = np.arange(len(labels))
ind
width = 0.6
fig, ax = plt.subplots(figsize=(6, 5))
rects = ax.barh(ind, np.array(values), color='red')
ax.set_yticks(ind+((width)/2.))
ax.set_yticklabels(labels, rotation='horizontal')
ax.set_xlabel("Cuenta de valores nulos")
ax.set_ylabel("Nombre de Columnas")
ax.set_title("Variables con valores nulos")
```

```python
# Imputación de variables nulas
from sklearn.impute import SimpleImputer
model = SimpleImputer(strategy='most_frequent')
titanic['embarked'] = model.fit_transform(
    titanic['embarked'].values.reshape(-1, 1))
titanic['embark_town'] = model.fit_transform(
    titanic['embark_town'].values.reshape(-1, 1))
print(titanic.isnull().sum())
```

```python
# imputación de variables nulas
titanic['age'] = titanic['age'].fillna(titanic['age'].mean())
print(titanic.isnull().sum())
```

```python
# Alternativa imputación con mmodelo de vecinoes cercanos
from sklearn.impute import KNNImputer
knn = KNNImputer(n_neighbors=1)
titanic['age'] = knn.fit_transform(titanic['age'].values.reshape(-1, 1))
print(titanic.isnull().sum())
```

```python
fig, ax = plt.subplots(figsize=(10, 5))
# ax = plt.axes()
x = np.linspace(0, 20, 30)
ax.plot(x, titanic.survived[:30], linestyle='dashed',
        marker='o', label='Sobrevivencia', color='orange')
ax.plot(x, titanic.embark_town[:30], linestyle='dashdot',
        marker='o', label='Puerto', color='green')
ax.set_xlim(0, 15)
ax.set_ylim(-1, 3)
ax.set_xlabel("Sobrevivencia")
ax.set_ylabel("Puertos")
ax.set_title("Puerto versus sobrevivencia")
ax.figure.legend()
```
