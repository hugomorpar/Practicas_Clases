## Integración de datos (merge)

```python
import pandas as pd 
ventas = pd.DataFrame({
    'producto': [1, 1,2,3,4, 5, 7, 6],
    'tienda': ['A', 'B', 'B', 'C', 'C', 'A', 'C', 'D'],
    'ventas': [100, 200, 50, 20, 20, 150, 250, 300]
})
   
producto = pd.DataFrame({
    'producto': [1,2,3,4,5,6,8],
    'descripcion': ['banano', 'pan blanco','pan integral', 'fideo', 'bebida', 'arroz', 'mango' ],
    'procedencia': ['ec', 'cl', 'cl', 'cl', 'ar', 'pe', 'pe']
}) 
```

```python
print(ventas.shape) 
print(producto.shape)
```

### Left Merge




```python
## realizar un LEFT merge de las tablas ventas y producto 

df0 = pd.merge(left= __________, right=___________ , on=________, how=_________)

display(df0)
```

### Right Merge 



```python
## realizar un RIGHT merge de las tablas ventas y producto 

df0 = pd.merge(left= ventas, right=_________ , on=___________, how='right')

display(df0)
```

### INNER merge 


```python
## realizar un RIGHT merge de las tablas ventas y producto 

df0 = pd.merge(left= ventas, right=producto , on='producto', how='inner')

display(df0)
```

### Transformación de una tabla : melt 

La función melt() en la biblioteca de Python llamada "pandas" se utiliza para transformar un DataFrame en un formato más largo y delgado, reorganizando los datos de las columnas en un nuevo DataFrame. Es especialmente útil cuando tienes un DataFrame con muchas columnas y deseas convertir esas columnas en filas, lo que puede facilitar el análisis y la visualización de los datos en algunos casos.

La función melt() toma varios parámetros, pero los principales son:

* id_vars: Una lista de columnas que se mantendrán sin cambios y se utilizarán como identificadores únicos para cada fila en el DataFrame resultante. Estas columnas se mantienen en su lugar mientras que las demás columnas se "derriten" o se transforman en filas.

* value_vars: Una lista de columnas que se van a convertir en valores dentro de una nueva columna llamada, por defecto, "value". Estas columnas se derriten y sus valores se colocan en la columna "value".

* var_name y value_name: Opcionales, permiten especificar los nombres de las columnas creadas en el DataFrame resultante para las columnas de variables y valores, respectivamente.


```python
# vamos aplicar un melt a la tabla producto  
display(producto) 
```

```python
producto.melt(id_vars=['producto'], value_vars=['descripcion', 'procedencia'])
```

```python
serie= pd.DataFrame({'trimestre':['T1', 'T2', 'T3', 'T4'],
                     '2020' : [85, 90, 105, 98],
                     '2021' : [100, 120, 150, 250], 
                     '2022' : [120, 230, 310, 280] })
display(serie)
```

```python
# realizar un melt, manteniendo la columna "trimestre"
# los años deben quedar bajo una columna llamada "año"
# los valores de las ventas bajo la columna "ventas"

df1= serie.melt(id_vars=['trimestre'], value_vars=['2020','2021', '2022'], value_name='venta', var_name='año')

display(df1)
```

## Handling Missing Data



```python
import numpy as np 

df = pd.DataFrame({'A':[1,2,np.nan],
                  'B':[5,np.nan,np.nan],
                  'C':[1,2,3]})
df
```

```python
# identificar si cada elemento es nulo o no 
df.isnull()
```

```python
df.notnull()
```

```python
## suma para todas las columnas la cantidad de nulos  
df.isnull().sum()
```

```python
## suma de nulos de la columna A 
df['A'].isnull().sum()
```

```python
## borra todas las filas que contengan al menos un nulo 
df.dropna()
```

```python
## reemplazando todos los nulos por un valor 

df.fillna(0)
```

```python
## reemplazando los nulos de una columna por su promedio  

df['A'].fillna(df['A'].mean())
```
