```python
# Farenheit to Celsius conversión
# (32°F − 32) × 5/9 = 0°C

import pandas as pd
import numpy as np

# ingreasmos la temperatura en farenheit con input

temp = float(input("Ingrese la temperatura en Farenheit: "))
celsius = (temp - 32) * 5/9
print("La temperatura en Celsius es: ", celsius)
```
