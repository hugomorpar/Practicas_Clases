# Explicacin

  Tipo: Las listas, tuplas, conjuntos y diccionarios de NumPy son objetos de diferentes clases.  
  
  Tamao: Las listas y tuplas de NumPy son de tamao fijo. Los conjuntos y diccionarios de NumPy son de tamao variable.  
  
  Tipo de elementos: Los elementos de las listas y tuplas de NumPy deben ser del mismo tipo. Los elementos de los conjuntos y diccionarios de NumPy pueden ser de cualquier tipo.  
  
  Acceso a elementos: Los elementos de las listas, tuplas y conjuntos de NumPy se pueden acceder mediante ndices. Los elementos de los diccionarios de
  
  NumPy se pueden acceder mediante key.  
  
  Operaciones matemticas: Las listas y tuplas de NumPy son eficientes para operaciones matemticas. Los conjuntos y diccionarios de NumPy son ineficientes para operaciones matemticas.  
  
  Operaciones de matriz: Las listas y tuplas de NumPy son eficientes para operaciones de matriz. Los conjuntos y diccionarios de NumPy son ineficientes para operaciones de matriz.  

  Memoria: Las listas y tuplas de NumPy ocupan ms memoria que los conjuntos y diccionarios de NumPy.  
  
  ## Uso con funciones de NumPy:

     Las listas y tuplas de NumPy son fciles de usar con funciones de NumPy. Los conjuntos y diccionarios de NumPy son difciles de usar con funciones de NumPy.  

  ## Sintaxis de cdigo:  
  
      Las listas de NumPy se crean usando la funcin numpy.array().  
      Las tuplas de NumPy se crean separando los elementos con comas y, opcionalmente, encerrndolos entre parntesis.  
      Los conjuntos de NumPy se crean usando la funcin set(). Los diccionarios de NumPy se crean usando la funcin dict().

Lista:  [ ] o list( )

```python
import numpy as np

X = [1, 2, 3, 4, 5]

print(X)

print( type(X))
```

Tupla: ( ) o tuple ( )

```python
import numpy as np

X = (1, 2, 3, 4, 5)

print(X)

print( type(X))
```

Conjunto: { } o set ( )

```python
import numpy as np

X = {1, 2, 3, 4, 5}

print (X)

print (type(X))
```

Diccionario: { } o dict ( )

```python
import numpy as np

X = {"nombre": "Juan", "edad": 25}

print(X)

print (type(X))
```

# DataFrame

```python
import pandas as pd

X = pd.DataFrame({'edad': [25, 30, 50], 'nombre': ['Juana', 'Alex', 'Norma']})

print( type(X))
```

```python
# display del DataFrame

display(X)
```

 La primera columna de la izquierda se llama "index" y siempre inica en cero.  
 Hay que considerarlo cuando se quiere acceder a un elemento de la tabla.

```python
print(X.shape)
```
