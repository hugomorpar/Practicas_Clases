## Diagrama de dispersin 

https://seaborn.pydata.org/generated/seaborn.scatterplot.html

```python
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime

hoy = datetime.now().strftime("%Y-%m-%d")

sns.scatterplot(data=df, x="Hours Studied",
                y="Performance Index", edgecolor="black", alpha=0.5)
plt.title("Diagrama de dispersin entre horas de estudio y desempeo ")
plt.suptitle("Elaborado por J.Tomal, Fecha: " +
             hoy+", Fuente: ...", fontsize=8, y=0)
```

## Diagrama de dispersin con lnea de tendencia 

```python
sns.lmplot(data=df, x="Hours Studied", y="Performance Index", scatter_kws={
           's': 5, 'alpha': 0.3}, line_kws={'color': 'red', 'alpha': 0.3})
plt.title("Diagrama de dispersin entre horas de estudio y desempeo ")
plt.suptitle("Elaborado por J.Tomal, Fecha: " +
             hoy+", Fuente: ...", fontsize=8, y=0)
```

## Calculando Correlacin 

```python
import seaborn as sn
import matplotlib.pyplot as plt
```



* La correlacin se calcula entre dos variables con escalas numricas
* La correlacin es un nmero entre -1 y 1 
* Una correlacin de 1 indica una relacin lineal perfecta y  directa
* Una correlacin de -1 indica una relacin lineal perfecta e inversa 
* Una correlacin de 0 indica que no existe una relacin lineal entre las variables 

```python
df['Sleep Hours'].corr(df['Hours Studied'])
```

```python
# df= df.drop([''])

corr_matrix = df.corr()
sns.heatmap(corr_matrix, annot=True)
plt.show()
```

```python
corr_matrix = df.corr()
sns.heatmap(corr_matrix, annot=True, cmap='Blues')
```
