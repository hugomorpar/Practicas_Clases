## Diagrama de dispersión 

https://seaborn.pydata.org/generated/seaborn.scatterplot.html

```python
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime

hoy = datetime.now().strftime("%Y-%m-%d")

sns.scatterplot(data=df, x="Hours Studied",
                y="Performance Index", edgecolor="black", alpha=0.5)
plt.title("Diagrama de dispersión entre horas de estudio y desempeño ")
plt.suptitle("Elaborado por J.Tomalá, Fecha: " +
             hoy+", Fuente: ...", fontsize=8, y=0)
```

## Diagrama de dispersión con línea de tendencia 

```python
sns.lmplot(data=df, x="Hours Studied", y="Performance Index", scatter_kws={
           's': 5, 'alpha': 0.3}, line_kws={'color': 'red', 'alpha': 0.3})
plt.title("Diagrama de dispersión entre horas de estudio y desempeño ")
plt.suptitle("Elaborado por J.Tomalá, Fecha: " +
             hoy+", Fuente: ...", fontsize=8, y=0)
```

## Calculando Correlación 

```python
import seaborn as sn
import matplotlib.pyplot as plt
```



* La correlación se calcula entre dos variables con escalas numéricas
* La correlación es un número entre -1 y 1 
* Una correlación de 1 indica una relación lineal perfecta y  directa
* Una correlación de -1 indica una relación lineal perfecta e inversa 
* Una correlación de 0 indica que no existe una relación lineal entre las variables 

```python
df['Sleep Hours'].corr(df['Hours Studied'])
```

```python
# df= df.drop([''])

corr_matrix = df.corr()
sns.heatmap(corr_matrix, annot=True)
plt.show()
```

```python
corr_matrix = df.corr()
sns.heatmap(corr_matrix, annot=True, cmap='Blues')
```
