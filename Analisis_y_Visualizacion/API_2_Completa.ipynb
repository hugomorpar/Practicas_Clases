# Dejando solo columnas de anio y poblacion_total

### Comenzamos a filtrar el DataFrame, dejando solo el campo de 'provincia', 'anio' y 'poblacion_total'

```python
# loc seleccion por label para ver solo las columnas de anio vs poblacion_total
df_1 = df.loc[:, ["anio", "poblacion_total"]]

display(df_1.head())
```

# Seleccionamos solo el anio 2010

```python
df_2 = df_1[df['anio'] == 2010]
df_2
```

# Sacamos la fila de "Total País"

```python
df_3 = df_2.drop(['Total País'])
# df_2 = df_1.drop(['0'])
```

# Importando los dataset 'hogares_viviendas_superficie.csv'

## Visualizamos el DataFrame y caracteristicas.

```python
import pandas as pd


# df = pd.read_csv("hogares_viviendas_superficie.csv")

df_4 = pd.read_csv('hogares_viviendas_superficie.csv',
                   index_col=0, encoding='latin-1')
```

```python
df_4.head(30)
```

```python
# Informacion de la tabla y Tipos de datos de las columnas (int64=integer de 64 bits)
df_4.info()
```

```python
# loc seleccion por label para ver solo las columnas de anio vs poblacion_total
df_5 = df_4.loc[:, ["provincia", "superficie_km2"]]

display(df_5.head(30))
```

# Igualando el 'index' en ambas tablas

```python
# Index utilizado: 'provincia'

df_6 = df_5.set_index('provincia')
```

# Concatenando ambas tablas

```python
# pegamos la columna 'poblacion_total' de df_3 en df_6

df_con = pd.merge(df_3, df_6, on='provincia')
```

```python
df_con
```

### Agregando una nueva columna con un campo nuevo "densidad" (habitantes/superficie)

```python
df_con['densidad'] = df_con['poblacion_total']/df_con['superficie_km2']
```
