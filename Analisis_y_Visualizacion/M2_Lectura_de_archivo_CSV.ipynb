# Seleccion de filas y columnas  
## iloc : utiliza los indices para la seleccion
## loc : utiliza las labels (etiquetas) para la seleccion

```python
# Seleccion de filas con index.
# Fila 3, 4 y 5 y todas las columnas (,:)

df2 = df.iloc[ [3, 4, 5], :]

display(df2.head())
```

```python
# Seleccion de columnas con index.
# Columnas con index 2 y 3

df2 = df.iloc[ : , [2,3]]

display(df2.head())
```

# Seleccionar columnas por label (etiqueta/nombre) con "loc" :




```python
df3 = df.loc[ : , ["Area code" , "International plan" , "Voice mail plan"] ]

display(df3.head())
```

# Filtros

```python
# Cuantos clientes tienen Total night minutes mayor a 100 ?

df[ df['Total night minutes']>100 ]
```

```python
# Filtrar los que corresponden al estado de OH o NY (usar isin)

df[df['State'].isin(['OH' , 'NY'])].shape
```

```python
# Seleccionar todos los registros que abandonaron la telefonica

df[df['Churn']==True]
```

```python
# Si queremos saber cuantos son, le agregamos el shape

df[df['Churn']==True].shape
```

```python
# Filtrar los clientes que abandonaron el servicio y tiene un "Total day charge" menor a 100

df[(df['Churn']==True) & (df['Total day charge']<100)]
```

```python
# Si queremos saber cuantos son, le agregamos el shape

df[(df['Churn']==True) & (df['Total day charge']<100)].shape
```

```python
# Calcular un campo nuevo

df['nueva_columna']= df['Total day charge']*2



display(df.head())
```

```python
# Calcular un campo nuevo

df['nueva_columna2']= df['Total day charge']/df['Total day calls']

display(df.head())
```
