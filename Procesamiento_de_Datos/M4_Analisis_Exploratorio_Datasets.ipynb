#  Análisis  Exploratorio  de  Datos  EDA  (Exploratory  Data  Analysis)

### Preguntas claves:
 * ¿Existe algún tipo de estructura en los datos?
 * ¿Dónde están centrados?
 * ¿Cuál es la dispersión?
 * ¿Cuál es la relación entre ellos?
 * ¿Cómo sintetizar y presentar la información?
 * ¿Hay sesgo en los datos seleccionados?
 * ¿Hay errores de codificación? ¿Hay datos atípicos?
 * ¿qué hacer con ellos? ¿Hay datos faltantes?
 * ¿Tiene algún patrón? ¿Qué hacer para recuperarlos?

### Usualmente EDA sigue una serie de pasos:

1. Familiarizarse con la naturaleza de los datos a analizar.
2. Realizar examen gráfico y descriptivo de las variables.
3. Realizar examen gráfico y descriptivo de la relación entre variables.
4. Preparar los datos para aplicar cualquier técnica.
5. Identifica atípicos y evalúa el impacto que estos provocan en el análisis.
6. Analiza el efecto de datos faltantes y toma decisiones.

## El objetivo principal de un EDA es detectar valores atipicos y errores

## normalized-losses, bore, stroke, horsepower, peak-rpm y price deberian ser del tipo 'integer' o 'float'. Aparecen como 'object' debido al simbolo '?'

## Una alternativa sería reemplazar los valores “?” por valor “nan” (es decir, nulo o vacío) y luego trabajar esos valores nulos reemplazándolos con la media o mediana.
## (esto se puede lograr en caso de que sean pocos en relación al total).

```python
# como resolverlo ?

for col in auto.columns:  # recorrer cada columna
    auto[col].replace({'?': np.nan}, inplace=True)  # reemplazar ? por NaN
auto.isnull().sum()  # verificar que se hayan reemplazado
```

```python
# Si quisiéramos visualizar la distribución de valores nulos, podemos usar un mapa de calor.

sns.heatmap(auto.isnull(), cbar=False, cmap='viridis')
```

```python
# Finalmente, podríamos rellenar los espacios vacíos para las variables numéricas con el valor de la media:

num_col = ['normalized-losses', 'bore',
           'stroke', 'horsepower', 'peak-rpm', 'price']
for col in num_col:
    auto[col] = pd.to_numeric(auto[col])
    auto[col].fillna(auto[col].mean(), inplace=True)
auto.head()
```

```python
fig = plt.figure()
ax = plt.axes()
x = np.linspace(0, 10, 1000)
ax.plot(x, np.sin(x))
ax.plot(x, np.cos(x))
```

```python
# ajustando el color de la grafica

# especificar el color por nombre
plt.plot(x, np.sin(x - 0), color='blue')
# codigo corto de color (rgbcmyk)
plt.plot(x, np.sin(x - 1), color='g')
plt.plot(x, np.sin(x - 2), color='0.75')        # Escala de grises entre 0 y 1
plt.plot(x, np.sin(x - 3), color='#FFDD44')     # Hex code (RRGGBB de 00 a FF)
# Tupla RGB, valores entre 0 y 1
plt.plot(x, np.sin(x - 4), color=(1.0, 0.2, 0.3))
# todos los nombres de colores HTML soportados
plt.plot(x, np.sin(x - 5), color='chartreuse')
```

```python
# definiendo el estilo de linea

plt.plot(x, x + 0, linestyle='solid')
plt.plot(x, x + 1, linestyle='dashed')
plt.plot(x, x + 2, linestyle='dashdot')
plt.plot(x, x + 3, linestyle='dotted')
```

```python
# abreviando el tipo de linea

plt.plot(x, x + 4, linestyle='-')  # solid
plt.plot(x, x + 5, linestyle='--')  # dashed
plt.plot(x, x + 6, linestyle='-.')  # dashdot
plt.plot(x, x + 7, linestyle=':')  # dotted
```

```python
# limite de los ejes

plt.plot(x, np.sin(x))
plt.xlim(-1, 11)
plt.ylim(-1.5, 1.5)
```

```python
# invertir los ejes

plt.plot(x, np.sin(x))
plt.xlim(10, 0)
plt.ylim(1.2, -1.2)
```

```python
# etiequetas de los ejes y titulo

plt.plot(x, np.sin(x))
plt.title("Una Curva Seno")
plt.xlabel("x")
plt.ylabel("sin(x)")
```

```python
#  ajustar  la  posición,  el  tamaño  y  el  estilo  de  estas  etiquetas

plt.plot(x, np.sin(x), '-g', label='seno(x)')
plt.plot(x, np.cos(x), ':b', label='coseno(x)')
plt.xlim(-1, 11)
plt.ylim(-1.5, 1.5)
plt.legend()
```
