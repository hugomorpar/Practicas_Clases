#  Análisis  Exploratorio  de  Datos  EDA  (Exploratory  Data  Analysis)

### Preguntas claves:
 * ¿Existe algún tipo de estructura en los datos?
 * ¿Dónde están centrados?
 * ¿Cuál es la dispersión?
 * ¿Cuál es la relación entre ellos?
 * ¿Cómo sintetizar y presentar la información?
 * ¿Hay sesgo en los datos seleccionados?
 * ¿Hay errores de codificación? ¿Hay datos atípicos?
 * ¿qué hacer con ellos? ¿Hay datos faltantes?
 * ¿Tiene algún patrón? ¿Qué hacer para recuperarlos?

### Usualmente EDA sigue una serie de pasos:

1. Familiarizarse con la naturaleza de los datos a analizar.
2. Realizar examen gráfico y descriptivo de las variables.
3. Realizar examen gráfico y descriptivo de la relación entre variables.
4. Preparar los datos para aplicar cualquier técnica.
5. Identifica atípicos y evalúa el impacto que estos provocan en el análisis.
6. Analiza el efecto de datos faltantes y toma decisiones.

## El objetivo principal de un EDA es detectar valores atipicos y errores

## normalized-losses, bore, stroke, horsepower, peak-rpm y price deberian ser del tipo 'integer' o 'float'. Aparecen como 'object' debido al simbolo '?'

## Una alternativa sería reemplazar los valores “?” por valor “nan” (es decir, nulo o vacío) y luego trabajar esos valores nulos reemplazándolos con la media o mediana.
## (esto se puede lograr en caso de que sean pocos en relación al total).

```python
# como resolverlo ?

for col in auto.columns:  # recorrer cada columna
    auto[col].replace({'?': np.nan}, inplace=True)  # reemplazar ? por NaN
auto.isnull().sum()  # verificar que se hayan reemplazado
```

```python
# Si quisiéramos visualizar la distribución de valores nulos, podemos usar un mapa de calor.

sns.heatmap(auto.isnull(), cbar=False, cmap='viridis')
```

```python
# Finalmente, podríamos rellenar los espacios vacíos para las variables numéricas con el valor de la media:

num_col = ['normalized-losses', 'bore',
           'stroke', 'horsepower', 'peak-rpm', 'price']
for col in num_col:
    auto[col] = pd.to_numeric(auto[col])
    auto[col].fillna(auto[col].mean(), inplace=True)
auto.head()
```

# Histogramas y boxplots

```python
# de barras discreto

import matplotlib.pyplot as plt
import numpy as np

eje_x = ['Carlos', 'Antonia', 'Maria', 'Juan']  # nombres de vendedores
eje_y = [50, 20, 35, 47]  # ventas en millones
plt.bar(eje_x, eje_y, color='lightgreen')  # graficar
plt.xlabel('Vendedores')  # nombre del eje x
plt.ylabel('Ventas en millones')  # nombre del eje y
plt.title('Ventas del mes por vendedor')  # titulo
```

```python
# este gráfico se podría invertir para visualizarlo horizontalmente con la función barh:

import matplotlib.pyplot as plt
import numpy as np

eje_x = ['Carlos', 'Antonia', 'Maria', 'Juan']  # nombres de vendedores
eje_y = [50, 20, 35, 47]  # ventas en millones
plt.barh(eje_x, eje_y, color='lightgreen')  # graficar horizontalmente
plt.xlabel('Ventas en millones')  # nombre del eje x
plt.ylabel('Vendedores')  # nombre del eje y
plt.title('Ventas del mes por vendedor')  # titulo
```

```python
# barras agrupadas

import matplotlib.pyplot as plt
import numpy as np

serie_1 = [406, 387, 442, 457, 485]  # ventas de la serie 1
serie_2 = [421, 453, 435, 478, 512]  # ventas de la serie 2

numero_de_grupos = len(serie_1)  # numero de grupos a graficar
indice_barras = np.arange(numero_de_grupos)  # indices de las barras
ancho_barras = 0.35  # ancho de las barras

plt.bar(indice_barras, serie_1, ancho_barras,
        label='Hombres')  # graficar serie 1
plt.bar(indice_barras + ancho_barras, serie_2,
        ancho_barras, label='Mujeres')  # graficar serie 2
plt.legend(loc='best')

# se colocan los indicadores en el eje x
plt.xticks(indice_barras + ancho_barras, ('2017', '2018',
           '2019', '2020', '2021'))  # etiquetas del eje x

plt.ylabel('Numero de habitantes')  # nombre del eje y
plt.xlabel('Años')  # nombre del eje x
plt.title('Numero de habitantes por genero'):  # titulo de la grafica
```

```python
# barras apiladas (distribucion de categorias)

import matplotlib.pyplot as plt
import numpy as np

grupos = ['Grupo 1', 'Grupo 2', 'Grupo 3', 'Grupo 4']  # nombres de los grupos
hombres = [28, 45, 34, 88]  # valores para hombres
mujeres = [10, 72, 37, 65]  # valores para mujeres
indice = np.arange(len(grupos))  # indice para los grupos

# se crean las primeras barras
plt.bar(indice, hombres, label='Hombres')

# se crean las segundas barras que se colocan sobre las primeras
# se colocan sobre las primeras
plt.bar(indice, mujeres, bottom=hombres, label='Mujeres')

plt.xticks(indice, grupos)  # se colocan los nombres de los grupos en el eje x
plt.ylabel('Participantes')  # nombre del eje y
plt.xlabel('Grupos')  # nombre del eje x
plt.title('Participantes por grupo y genero')  # titulo de la grafica
plt.legend(loc=2)  # ubicacion donde se coloca la leyenda
```

```python
# histograma

# Ilustra la frecuencia absoluta o relativa en datos numéricos continuos

%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

print(plt.style.available)  # estilos disponibles

plt.style.use('seaborn-v0_8-white') # estilo de la grafica
data = np.random.randn(1000)  # datos aleatorios
plt.hist(data, bins=10);  # graficar histograma, por defecto 10 bins, plt.hist(data, bins=20)
```

```python
# otros atributos de hist()

# data: datos a graficar
# bins: numero de divisiones o contenedores
# alpha: controla la transparencia de las barras
# histtype: tipo de histograma (bar, barstacked, step, stepfilled)
# color: color de las barras
# edgecolor: color del borde de las barras

plt.hist(data, bins=30, alpha=0.5, histtype='stepfilled',
         color='steelblue', edgecolor='black')  # graficar histograma
```

```python
import matplotlib.pyplot as plt
import numpy as np

x1 = np.random.normal(0, 0.8, 1000)  # datos aleatorios
x2 = np.random.normal(-2, 1, 1000)  # datos aleatorios
x3 = np.random.normal(3, 2, 1000)  # datos aleatorios
kwargs = dict(histtype='stepfilled', alpha=0.3, bins=40)  # parametros
plt.hist(x1, **kwargs)  # graficar histograma
plt.hist(x2, **kwargs)  # graficar histograma
plt.hist(x3, **kwargs)  # graficar histograma
```
